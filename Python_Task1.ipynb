### Q1. Create a function which will take a list as an argument and return the product of all the numbers after creating a flat list.
Use the below given list as an argumnet for your fuction.
```
list1 = [1,2,3,4, [44,55,66, True], False, (34,56,78,89,34), {1,2,3,3,2,1}, {1:34, "key2": [55, 67, 78, 89], 4: (45,
22, 61, 34)}, [56, 'data science'], 'Machine Learning']
```
Note : You must extract numeric keys and values of the dictionary also.

- **Step-By-Step Approach**
  - Create an empty list to store numbers.
  - Traverse each element of the lsit.
  - Check type of element
  - If element is:
    - list/tuple/set -> iterate inside
    - dictionary -> extract keys and values
  - Append numeric values to list
  - Multiply all numbers.

````python
# Python Implementation

from functools import reduce
import operator

def extract_numbers(data):
  numbers = []

  for item in data:

    # if item is integer or boolean
    if isinstance(item, (int, bool)):
      numbers.append(int(item))

    # if item is list, tuple, or set
    elif isinstance(item, (list, tuple, set)):
      for i in item:
        if isinstance(i, (int, bool)):
          numbers.append(int(i))

    # if item is dictionary
    elif isinstance(item, dict):

      # extract keys
      for key in item.keys():
        if isinstance(key, (int, bool)):
          numbers.append(int(key))

      # extract values
      for value in item.values():

        if isinstance(value, (int, bool)):
          numbers.append(int(value))

        elif isinstance(value, (list, tuple)):
          for v in value:
            if isinstance(v, (int, bool)):
              numbers.append(int(v))

  return numbers

def product_of_numbers(lst):

  nums = extract_numbers(lst)
  product = reduce(operator.mul, nums, 1)
  return product

list1 = [1,2,3,4,[44,55,66,True],False,(34,56,78,89,34),{1,2,3,3,2,1},{1:34,"key2":[55,67,78,89],4:(45,22,61,34)},[56,'data science'],'Machine Learning']

result = product_of_numbers(list1)
print(result)
````

### Q2. Write a python program for encrypting a message sent to you by your friend. The logic of encryption should be such that, for a the output should be z. For b, the output should be y. For c, the output should be x respectively. Also, the whitespace should be replaced with a dollar sign. Keep the punctuation marks unchanged.

Input sentence: I want to become a Data Scientist.

Encrypt the above input sentence using the program you just created.

Note : Convert the given input sentence into lowercase before encrypting. The final output should be lowercase.

- The given encryption is reverse alphabet encryption
- also:
  - convert text to lowercase
  - replace whitespace with $
  - keep punctuation unchanged

````python
# Python Implementation

def encrypt_message(sentence):

    sentence = sentence.lower()

    encrypted = ""

    for char in sentence:

        if char.isalpha():
            new_char = chr(ord('z') - (ord(char) - ord('a')))
            encrypted += new_char

        elif char == " ":
            encrypted += "$"

        else:
            encrypted += char

    return encrypted


sentence = "I want to become a Data Scientist."

result = encrypt_message(sentence)

print("Original Sentence:")
print(sentence)

print("\nEncrypted Sentence:")
print(result)
````

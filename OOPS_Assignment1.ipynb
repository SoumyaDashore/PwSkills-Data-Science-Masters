### Q1. Explain class and object with respect to object- oriented programming. Give a suitable example.

- **class**
  - A class is a blueprint or template used to create objects. It defines the properties(variables) and behaviours(methods/functions) that the objects created from the class will have.
  - In simple terms:
    - Class = Blueprint
    - defines attributes and methods
    - does not occupy memory untill objects are created.
- **Object**
  - An object is an instance of a class. When a class is created, it only defines the structure. When we create an object, memory is allocated and the object can use the class properties and methods.

````python
# Defining a class
class Car:

    # Constructor
    def __init__(self, brand, color):
        self.brand = brand
        self.color = color

    # Method
    def display_info(self):
        print("Car Brand:", self.brand)
        print("Car Color:", self.color)


# Creating objects
car1 = Car("Toyota", "Red")
car2 = Car("Tesla", "Black")

# Calling methods
car1.display_info()
print()
car2.display_info()
````

### Q2. Name the four pillars of OOPs.

- Object- Oriented Programming is based on four main principles(pillars).
- **1. Encapsulation**
  - Encapsulation means wrapping data and methods together into a single unit(class) and restricting direct access to some components.
- **2. Inheritance**
  - Inheritance allows a child class to inherit properties and methods from a parent class.
- **3. Polymorphism**
  - Polymorphism means same method name but different behaviour.
- **4. Abstraction**
  - Abstraction means hiding complex implementation and showing only essential features.

````python
# Example for Encapsulation

class BankAccount:

  def __init__(self, balance):
    self.__balance = balance

  def deposit(self, amount):
    self.__balance += amount

  def get_balance(self):
    return self.__balance

account = BankAccount(1000)
account.deposit(500)
print(account.get_balance())
````

````python
# Example for Inheritance

class Animal:

    def speak(self):
        print("Animal makes sound")

class Dog(Animal):

    def bark(self):
        print("Dog barks")

d = Dog()
d.speak()
d.bark()
````

````python
# Example for Polymorphism

class Bird:

    def sound(self):
        print("Bird makes sound")

class Parrot(Bird):

    def sound(self):
        print("Parrot talks")

class Crow(Bird):

    def sound(self):
        print("Crow caws")

p = Parrot()
c = Crow()

p.sound()
c.sound()
````

````python
# Example for Abstraction

from abc import ABC, abstractmethod

class Shape(ABC):

    @abstractmethod
    def area(self):
        pass

class Circle(Shape):

    def __init__(self, r):
        self.r = r

    def area(self):
        return 3.14 * self.r * self.r

c = Circle(5)
print("Area:", c.area())
````

### Q3. Explain why the __init__() function is used. Give a suitable example.

- Definition
  - `__init__()` is a constructor method in python. It is automatically called when an object of a class is created.
- Purpose
  - It is used to initialize object attributes
  - used to assign values to variables
  - used to perform setup operations
- Syntax
```
def __init__(self):
```

````python
# Example of __init__() function

class Student:

    def __init__(self, name, age):
        self.name = name
        self.age = age

    def display(self):
        print("Name:", self.name)
        print("Age:", self.age)


# Object creation
s1 = Student("Rahul", 20)
s2 = Student("Anita", 21)

s1.display()
print()
s2.display()
````

### Q4. Why self is used in OOPs?

- self represents the current object of the class.
- It is used to:
  - 1. access instance variables
  - 2. access methods of the same class
  - 3. differentiate between class variables and local variables
- self is not a keyword, but a naming convention.
- It must be the first parameter in class methods

````python
# Example for self usage

class Person:

    def __init__(self, name):
        self.name = name

    def greet(self):
        print("Hello, my name is", self.name)


p1 = Person("Rohan")
p1.greet()
````

### Q5. What is inheritance? Give an example for each type of inheritance.

- **Definition**
  - Inheritance is the mechanism in which one class acquires properties and methods from another class.
    - parent class -> base class
    - child class -> derived class
  - benefits:
    - code resuability
    - reduces redundancy
    - improves maintainability
- **Types of Inheritance in Python**
  - **1. Single Inheritance**
    - One child class inherits from one parent class.
  - **2. Multiple Inheritance**
    - A child class inherits from multiple parent class.
  - **3. Multilevel Inheritance**
    - Inheritance occurs in multiple levels.
  - **4. Hierarchial Inheritance**
    - Multiple child classes inherit from a single parent class.
  - **5. Hybrid Inheritance**
    - Combination of two or more types of inheritance.

````python
# Example for Single Inheritance

class Parent:

    def show(self):
        print("This is parent class")

class Child(Parent):

    def display(self):
        print("This is child class")

c = Child()
c.show()
c.display()
````

````python
# Example for Multiple Inheritance

class Father:

    def skill1(self):
        print("Father: Driving")

class Mother:

    def skill2(self):
        print("Mother: Cooking")

class Child(Father, Mother):
    pass

c = Child()

c.skill1()
c.skill2()
````

````python
# Example for Multilevel Inheritance

class Grandparent:

    def house(self):
        print("Grandparent's House")

class Parent(Grandparent):

    def car(self):
        print("Parent's Car")

class Child(Parent):

    def bike(self):
        print("Child's Bike")

c = Child()

c.house()
c.car()
c.bike()
````

````python
# Example for Hierarchial Inheritance

class Animal:

    def eat(self):
        print("Animal eats food")

class Dog(Animal):

    def bark(self):
        print("Dog barks")

class Cat(Animal):

    def meow(self):
        print("Cat meows")

d = Dog()
c = Cat()

d.eat()
d.bark()

c.eat()
c.meow()
````

````python
# Example for Hybrid Inheritance

class A:
    def methodA(self):
        print("Class A")

class B(A):
    def methodB(self):
        print("Class B")

class C(A):
    def methodC(self):
        print("Class C")

class D(B, C):
    def methodD(self):
        print("Class D")

d = D()

d.methodA()
d.methodB()
d.methodC()
d.methodD()
````

### Q1. What are the characteristics of the tuples? Is tuple immutable?
- Characteristics of Tuples:
  - Ordered - Elements have a fixed order.
  - Immutable - Once created, elements cannot be modified, added, or removed.
  - Allows Duplicates - Duplicate values are allowed.
  - Heterogeneous Data Types - Can store different data types in a single tuple.
  - Indexed - supports indexing and slicing.
- Yes, Tuples are immutable.
  - Once a tuple is created, you cannot modify its elements.
  - Example
  ```
  # Creating a tuple
  tup = (1,2,3,4)

  # Trying to modify
  tup[0] = 10                # This will raise TypeError

  # Output - TypeError: 'tuple' object does not support item assignment
  ```

### Q2.  What are the two tuple methods in python? Give an example of each method. Give a reason why tuples have only two in-built methods as compared to Lists.
- Tuples have only two built-in methods:
  - count()
  - index()
- 1. count()
  - Returns number of times a value appears
  ```
  tup = (1, 2, 3, 2, 2, 4)

  print(tup.count(2))
  ```
  - Output
  ```
  3
  ```
- 2. index()
  - Returns index of first occurrence.
  ```
  tup = (10,20,30,40)

  print(tup.index(30))
  ```
  - Output
  ```
  2
  ```
- Only two methods because tuples are immutable. They cannot be modified, so methods like:
  - append()
  - remove()
  - pop()
  - extend()
   
   are not required.

### Q4. Explain the difference between the union() and update() methods for a set. Give an example of each method.
- Union() method.
  - It returns a new set.
  - It does not modifies the original set.
  - Example of union()
  ```
  A = {1,2,3}
  B = {3,4,5}

  C = A.union(B)

  print("A:", A)
  print("union result:", C)
  ```
  - Output
  ```
  A: {1,2,3}
  Union result: {1,2,3,4,5}
  ```
- Update() method.
  - It does not returns a new set.
  - It modifies the original set.
  - Example
  ```
  A = {1,2,3}
  B = {3,4,5}

  A.update(B)

  print("Updated A:", A)
  ```
  - Output
  ```
  Updated A: {1,2,3,4,5}
  ```

### Q6.  Can we create a nested dictionary? If so, please give an example by creating a simple one-level nested dictionary.

### Q8. What are the three view objects in dictionaries? Use the three in-built methods in python to display these three view objects for the given dictionary.
```
 dict1 = {'Sport': 'Cricket' , 'Teams': ['India', 'Australia', 'England', 'South Africa', 'Sri Lanka', 'New Zealand']}
 ```
 - Three view objects in dictionary are:
   - keys()
   - values()
   - items()

````python
dict1 = {
    'Sport': 'Cricket',
    'Teams': ['India', 'Australia', 'England', 'South Africa', 'Sri Lanka', 'New Zealand']
}

# Keys view
print("Keys:", dict1.keys())

# Values view
print("Values:", dict1.values())

# Items view
print("Items:", dict1.items())
````

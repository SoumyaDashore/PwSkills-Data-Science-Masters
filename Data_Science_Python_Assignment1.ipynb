### Q2. Which type of Programming does Python support?

- Python supports multiple programming paradigms, including:
- **1. Procedural Programming**
  - Program is written as a sequence of functions.
  ```
  def greet():
    print("Hello World")

  greet()
  ```
- **2. Object-Oriented Programming**
  - Python supports classes and objects
  ```
  class Student:
    def __init__(self, name):
       self.name = name
  
  s = Student("Rahul")
  print(s.name)
  ```
- **3. Functional Programming**
  - Python supports functions as first- class objects.
  ```
  square = lambda x: x*x
  print(square(5))
  ```
- **4. Scripting**
  - Python is widely used for automation scripts.
  ```
  import os
  print(os.getword())
  ```

### Q5. Is python code compiled or interpreted?

- Python is an compiled-interpreted language.
- However, internally it works in two steps:
  - 1. Python code -> Bytecode compilation
  - 2. Bytecode -> Executed by python virtual machine(PVM)
- Execution Process
```
Source Code (.py)
        ↓
Bytecode (.pyc)
        ↓
Python Virtual Machine
        ↓
Output
```

### Q9. Python supports the creation of anonymous functions at runtime, using a construct called ____.

- Python uses lambda functions to create anonymous functions.
- **Lambda Functions**
  - A lambda function is a small anonymous function without a name.
  - Syntax:
  ```
  lambda arguments : expression
  ```

````python
# Example of lambda function

square = lambda x: x*x
print(square(5))
````

### Q14. How is memory managed in python?

- Python uses Automatic Memory Management
- It involves:

  **1. Private Heap**
  
  All python objects are stored in a private heap memory.

  **2. Reference Counting**

  Python tracks the number of references to an object.

  **3. Garbage Collection**

  When objects are no longer used, python automatically deletes them.

### Q16. Is indentation required in python?

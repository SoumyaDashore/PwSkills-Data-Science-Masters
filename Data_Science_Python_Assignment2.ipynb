### Q1. Write a program to accept percentage from the user and display the grade according to the following criteria:

|    Marks      |    Grade      |
|:------------- |:------------- |
|    >90        |     A         |
|> 80 and <= 90 |     B         |
|>= 60 and <= 80|     C         |
|    below 60   |     D         |

````python
# Solution for Q1

try:
  percentage = float(input("Enter your percentage: "))

  if percentage > 90:
    print("Grade: A")

  elif percentage > 80 and percentage <= 90:
    print("Grade: B")

  elif percentage >= 60 and percentage <= 80:
    print("Grade: C")

  else:
    print("Grade: D")

except ValueError:
  print("Invalid Input! Please enter numeric value.")
````

### Q2. Write a program to accept cost price of a bike and display the road tax to be paid according to the following criteria:

|    Task   |  Cost Price(in RS)  |
|:----------|:------------------- |
|    15%    |     >100000         |
|    10%    |>50000 and <= 100000 |
|    5%     |     <= 50000        |


````python
# Solution for Q2.

try:
  cost_price = float(input("Enter the cost price of the bike: "))

  if cost_price > 100000:
    road_tax = cost_price * 0.15

  elif cost_price > 50000 and cost_price <= 100000:
    road_tax = cost_price * 0.10

  else:
    road_tax = cost_price * 0.05

  print(f"Road Tax to be paid: {road_tax}")

except ValueError:
  print("Please enter a valid number")
````

### Q3. Accept any city from the user and display monuments of that city.

|    City       |     Monument  |
|:------------- |:------------- |
|    Delhi      |    Red Fort   |
|    Agra       |    Taj Mahal  |
|    Jaipur     |    Jal Mahal  |

````python
# Solution for Q3.

city = input("Enter city name: ").lower()

if city == "delhi":
  print("Monument: Red Fort")

elif city == "agra":
  print("Monument: Taj Mahal")

elif city == "jaipur":
  print("Monument: Jal Mahal")

else:
  print("City not in database")
````

### Q4. Check how many times a given number can be divided by 3 before it is less than or equal to 10.

````python
# Solution for Q4.

number = int(input("Enter a number: "))
count = 0

while number > 10:
  number = number / 3
  count += 1

print(f"Number of times divisible by 3: {count}")
````

### Q5. Why and When to use while loop in python give a detailed description with example.

- What is While loop?
  - A while loop repeatedly executes a block of code as long as condition is true.
  - syntax
  ```
  while condition:
    statement
  ```
- When to use while loop
  - when number of iterations is not known beforehand
  - loop depends on a condition
  - input-based repetition
  - real-time processing

````python
# Example of while loop

num = 1

while num <= 5:
  print(num)
  num += 1
````

### Q6. Use nested while loop to print 3 different pattern.

````python
# Pattern 1

i = 1
while i <= 5:
  j = 1
  while j <= i:
    print("*", end="")
    j += 1
  print()
  i += 1
````

````python
# Pattern 2

i = 1

while i <= 5:
    j = 1
    while j <= i:
        print(j, end="")
        j += 1
    print()
    i += 1
````

````python
# Pattern 3

i = 5

while i >= 1:
    j = 1
    while j <= i:
        print("*", end="")
        j += 1
    print()
    i -= 1
````

### Q7. Reverse a while loop to display numbers from 10 to 1.

````python
# Solution for Q7.

num = 10

while num >= 1:
  print(num)
  num -= 1
````

### Q8. Reverse a while loop to display numbers from 10 to 1.(Alternative way than previous Question)

````python
# Solution for Q8.

num = 10

while True:
  print(num)
  num -= 1

  if num == 0:
    break
````
